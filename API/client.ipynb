```python
# client.py

import requests

base_url = "http://localhost:8001"

def test_predict_full():
    data = {
        "Kilometrage": 150248,
        "Annee_Facture": 2021,
        "Annee_veh": 2017,
        "Marque_veh": "PEUGEOT",
        "Modele_veh": "208 I Ph2",
        "Type_Energie": "Thermique",
        "Carburant": "Diesel"
    }

    response = requests.post(f"{base_url}/predict/full", json=data)

    if response.status_code == 200:
        print("Réponse du modèle complet :", response.json())
    else:
        print("Erreur pour le modèle complet :", response.status_code, response.text)

if __name__ == "__main__":
    test_predict_full()
```
